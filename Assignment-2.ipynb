```python
import sympy as sy
from sympy import *
```

```python
import random
```

```python
#step1
X=random.randint(1,5)
Y=random.randint(1,5)
print("random values of x:{}".format(X)+"random values of Y:{}".format(Y))
n=0.01
epochs=100
itr=0
x=0
x=sy.Symbol('x')
y=sy.Symbol('y')
print("function f=a*x**d+b*y**e+c")
a=int(input("enter the coefficient of x"))
d=int(input("enter the power of x"))
b=int(input("enter the coefficient of y"))
e=int(input("enter the power of y"))
c=int(input("enter the constant value"))
f=a*x**d+b*y**e+c
print("Input function f(x,y):{}".format(f))
print("initialization is done the function is {}".format(f))
while 1:
    print("the iteration is:{}".format(itr))
    #step-2
    dfx=sy.diff(f,x)
    dfy=sy.diff(f,y)
    slopex = dfx.subs(x,X) #after substitution of X in slope (dfx)
    slopey = dfy.subs(y,Y) #after substitution of Y in slope (dfy)
    print("slope(x):{}".format(slopex)+"slope(y):{}".format(slopey))
    #step-3
    delta_X=-n*slopex
    delta_Y=n*slopey
    print("delta_X:{}".format(delta_X)+"delta_Y:{}".format(delta_Y))
    #step-4
    X=X+delta_X
    Y=Y+delta_Y
    print("updated value of X:{}".format(X)+"updated value of Y:{}".format(Y))
    resf=f.subs(x,X)
    res=resf.subs(y,Y)
    print("result value of the function f(updated_X,updated_Y):{}".format(res))
    #step-5:update itr value
    print("step-5:update iteration")
    itr=itr+1
    #step-7
    if(itr>epochs):
        break;
    else:
        print("go to step-2")
    print("---------------------------------------------------------------------------")

print("***************************************************************************")
print("Global Minima point obtained: after rounding to two values after decimal point")
print("X: ",round(X,2))
print("Y: ",round(Y,2))
optimalres = f.subs(x,X)
optimalfinal = optimalres.subs(y,Y)
print("Optimal function value: ",round(optimalfinal,2))


```
